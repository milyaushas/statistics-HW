```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#U[0, 1]

n = 1000
maxK = 100
maxS = 1000

def getSD(k):
    SD = 0
    for s in range(1, maxS + 1):
        sample = np.random.uniform(0, 1, n)
        m_k = sum([elem**k for elem in sample]) / n
        theta = (m_k * (k + 1))**(1/k)
        SD += (theta - 1) ** 2
    return SD / s

k = [i for i in range(1, maxK + 1)]
SDs = list(map(getSD, k))

plt.plot(k, SDs, label = "uniform distribution")

plt.xlabel("k")
plt.ylabel("standart deviation")
plt.grid()
plt.legend()
```

Для равномерного распределения СКО уменьшается при росте k, поэтому для оценки параметра стоит брать большие k, но даже при k > 15 мы уже получаем значения СКО меньше 0.00005

```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

#Exp(1)

n = 1000
maxK = 100
maxS = 1000

def getSD(k):
    SD = 0
    for s in range(1, maxS + 1):
        sample = np.random.exponential(1, n)
        m_k = sum([elem**k for elem in sample]) / n
        theta = (m_k / math.factorial(k))**(1/k)
        SD += (theta - 1) ** 2
    return SD / s

k = [i for i in range(1, maxK + 1)]
SDs = list(map(getSD, k))

plt.plot(k, SDs, label = "exponential distribution")

plt.xlabel("k")
plt.ylabel("standart deviation")
plt.grid()
plt.legend()
```

Для экспоненциального же, наоборот, функция на графике монотонно возрастает, поэтому нужно использовать маленькие k < 5
